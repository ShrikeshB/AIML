```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./Resources/Salaries.csv")
```

```python
sns.histplot(data=data,x='salary', kde=True)
plt.show()
```

```python
sns.boxplot(data=data,x='salary')
plt.show()
```

```python
sns.boxplot(data=data,y='salary')
plt.show()
```

```python
sns.countplot(data=data,x='rank')
plt.show()
```

```python
data.head()
```

```python
sns.lineplot(data=data, x=data.index,y='salary')
plt.show()
```

```python
sns.pointplot(data=data,x=data.index,y='salary')
```

```python
sns.relplot(data=data, x='yrs.service',y='salary')
plt.show()
```

```python
sns.relplot(data=data, x='yrs.service',y='salary',hue='gender')
plt.show()
```

```python
sns.relplot(data=data, x='yrs.service',y='salary', hue="gender",style='rank')
```

### regression plot

```python
sns.regplot(data=data,x='yrs.service',y='salary')
```

### Barplot

```python
sns.barplot(data=data,x='rank',y='salary')
plt.show()
```

```python
sns.swarmplot(data=data,x='gender',y='salary')
```

```python
# sns.stripplot(data=data,x='gender',y='salary')
sns.stripplot(data=data,x='gender',y='salary',jitter=True)
```

```python
sns.boxplot(data=data,x='gender',y='salary',color='lightgray')
sns.stripplot(data=data,x='gender',y='salary',jitter=True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(data=data,x='gender',y='salary')
plt.savefig('violinplot.png')
plt.show()
```

```python
sns.pairplot(data=data, hue='rank')
```

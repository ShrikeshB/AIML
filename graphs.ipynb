```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = [1,1,3,4,4]
y = [5,10,15,19,20]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel("X VALUES")
plt.ylabel("Y VALUES")
plt.title("GRAPHS")
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel("X VALUES", fontsize=14, color='red')
plt.ylabel("Y VALUES", fontsize=15, color='blue')
plt.title("GRAPHS", fontsize=22, color='black')
plt.plot(x,y,marker='o', markersize=10, markerfacecolor='red', markeredgecolor='black',
         linestyle='--', linewidth=2, label="sample line")
plt.show()
```

### graph for showing features

```python
plt.xlabel("X VALUES", fontsize=14, color='black')
plt.ylabel("Y VALUES", fontsize=15, color="black")
plt.title("BAR CHART", fontsize=22, color='black')
plt.bar(x,y,color='r',width=0.25, edgecolor='grey')
plt.show()
```

```python
sections = ['A','B','C','D']
TotalStrength = [45,50,80,25]
```

```python
plt.xlabel("SECTIONs", fontsize=14, color='black')
plt.ylabel("strength", fontsize=15, color='black')
plt.title("BAR CHART", fontsize=15, color='red')
plt.bar(sections,TotalStrength, color='r', edgecolor='black')
# plt.plot(sections,TotalStrength,marker='o', markersize=10, markerfacecolor='red', markeredgecolor='black',
#          linestyle='--', linewidth=2, label="sample line")
plt.show()
```

```python
plt.savefig('bar_plot.pdf')
plt.show()
```

```python
plt.xlabel("SECTIONs", fontsize=14, color='black')
plt.ylabel("strength", fontsize=15, color='black')
plt.title("BAR CHART", fontsize=15, color='red')
plt.barh(sections,TotalStrength, color='r', edgecolor='black')
# plt.plot(sections,TotalStrength,marker='o', markersize=10, markerfacecolor='red', markeredgecolor='black',
#          linestyle='--', linewidth=2, label="sample line")
plt.show()
```

### histogram

```python
plt.xlabel('X VALUES', fontweight='bold',fontsize=14, color='black')
plt.xlabel('Y VALUES', fontweight='bold',fontsize=14, color='black')
plt.title("histogram", fontsize=14, color='red')
plt.hist(x, color='g', width=0.25,edgecolor='black')
plt.show()
```

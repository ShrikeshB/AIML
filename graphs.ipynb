### PIE CHART

```python
subjects = ['C','JAVA','C++','C#']
students = [23,12,52,51]
```

```python
plt.title('pie chart', color='red')
colors = ['gold','yellowgreen','lightcoral','lightskyblue','maroon']
plt.pie(students,labels=subjects,colors=colors,autopct='%1.1f%%')
plt.show()
```

```python
plt.title('pie chart', color='red')
colors = ['gold','yellowgreen','lightcoral','lightskyblue','maroon']
plt.pie(students,labels=subjects,colors=colors,autopct='%1.1f%%' , radius=0.9)
plt.legend(subjects,loc='lower right')
plt.show()
```

### Scatter plot

```python
plt.xlabel('X VALUES',fontsize=14, color='black')
plt.xlabel('Y VALUES',fontsize=14, color='black')
plt.title("scatter", fontsize=14, color='red')
plt.scatter(x,y)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 1000
colors = np.random.rand(50)
plt.scatter(x,y,s=sizes,c=colors,alpha=0.5, 
            cmap='viridis', edgecolor='black', linewidth=1.5,marker='o',label='data point')
```

### Area Chart

```python
x = np.linspace(0,10,100)
y=np.sin(x)
plt.fill_between(x,y,color='skyblue',alpha=0.4)
plt.plot(x,y,color='black',alpha=1, linewidth=2)
plt.title("AREA CHART")
plt.show()
```

```python
x1= [1,1,3,4,4]
y1=[5,13,24,5,21]
plt.fill_between(x1,y1, color='blue',alpha=0.6)
plt.plot(x1,y1,color='black',alpha=1,linewidth=2)
plt.title("area chart")
plt.show()
```

```python
np.random.seed(0)
x2 = np.random.randn(1000)
y2 = np.random.randn(1000)
plt.hexbin(x2,y2,gridsize=30,cmap='Blues')
plt.colorbar(label='count')
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.title("hexa bin plt for feature exploration")
plt.show()
```
